```python
# Variables seem to have incorrect type and inconsistent formatting. You need to fix them: 
def convert_size(Size):
    if isinstance(Size, str):
        if 'k' in Size.lower() or 'k' in Size:
            return float(Size.replace('k', '').replace('K', '')) / 1024  # Convert KB to MB
        elif 'm' in Size.lower():
            return float(Size.replace('M', '').replace('m', ''))
        elif Size == 'Varies with device':
            return 'nan'
    return Size
```

```python
def clean_reviews_installs(df):
    # Convert to numeric, forcing non-numeric to NaN
    df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
    df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
    
    # Drop rows where either column is NaN
    initial_count = len(df)
    df = df.dropna(subset=['Reviews', 'Installs'])
    print(f"Dropped {initial_count - len(df)} rows with missing values")
    
    # Find and remove invalid records
    invalid = df[df['Reviews'] > df['Installs']]
    print(f"Found {len(invalid)} records with reviews > installs")
    
    if len(invalid) > 0:
        print("\nSample of invalid records:")
        print(invalid[['App', 'Reviews', 'Installs']].head())
    
    df = df[df['Reviews'] <= df['Installs']]
    return df

df_clean = clean_reviews_installs(df_clean)
```

```python
invalid_prices = df_clean[(df_clean['Type'] == 'Free') & (df_clean['Price'] > 0)]
print(f"Free apps with price > 0: {len(invalid_prices)}")
df_clean = df_clean[~((df_clean['Type'] == 'Free') & (df_clean['Price'] > 0))]
```

```python
def clean_and_plot_size_vs_rating(df):
    # Convert to numeric
    df['Size'] = pd.to_numeric(df['Size'], errors='coerce')
    df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
    
    # Remove invalid/missing data
    plot_data = df.dropna(subset=['Size', 'Rating'])
    
    # Check data ranges
    print(f"Size range: {plot_data['Size'].min():.2f} to {plot_data['Size'].max():.2f} MB")
    print(f"Rating range: {plot_data['Rating'].min():.2f} to {plot_data['Rating'].max():.2f}")
    
    # Create plot
    plt.figure(figsize=(10,6))
    sns.jointplot(
        x='Size', 
        y='Rating', 
        data=plot_data, 
        kind='hex',
        xlim=(0, plot_data['Size'].quantile(0.95)),  # Show up to 95th percentile
        ylim=(1, 5)  # Rating scale
    )
    plt.suptitle('Rating vs. App Size (MB)', y=1.02)
    plt.show()
    
    return plot_data

cleaned_data = clean_and_plot_size_vs_rating(df_clean)
```
